```python
#1. Importamos todos los datos
#Leer datos .mat
import scipy.io 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns


#Notacion exponencial
pd.set_option('display.float_format', '{:e}'.format)

ha1 = scipy.io.loadmat('H-A-1.mat')
ha2 = scipy.io.loadmat('H-A-2.mat')
ha3 = scipy.io.loadmat('H-A-3.mat')

ia1 = scipy.io.loadmat('I-A-1.mat')
ia2 = scipy.io.loadmat('I-A-2.mat')
ia3 = scipy.io.loadmat('I-A-3.mat')


oa1 = scipy.io.loadmat('O-A-1.mat')
oa2 = scipy.io.loadmat('O-A-2.mat')
oa3 = scipy.io.loadmat('O-A-3.mat')



sorted(ha1.keys())
sorted(ha2.keys())
sorted(ha3.keys())

sorted(ia1.keys())
sorted(ia2.keys())
sorted(ia3.keys())

sorted(oa1.keys())
sorted(oa2.keys())
sorted(oa3.keys())

xh1_1 = ha1['Channel_1'] #Muestra los datos primer canal
xh1_2 = ha2['Channel_1']
xh1_3 = ha3['Channel_1']

xi1_1 = ia1['Channel_1']
xi1_2 = ia2['Channel_1']
xi1_3 = ia3['Channel_1']

xo1_1 = oa1['Channel_1']
xo1_2 = oa2['Channel_1'] 
xo1_3 = oa3['Channel_1'] 



n=len(xo1_1)

canalh1_1 = np.zeros(n)
canalh1_2 = np.zeros(n)
canalh1_3 = np.zeros(n)

canali1_1 = np.zeros(n)
canali1_2 = np.zeros(n)
canali1_3 = np.zeros(n)

canalo1_1 = np.zeros(n)
canalo1_2 = np.zeros(n)
canalo1_3 = np.zeros(n)


for i in range(0,n):
    canalh1_1[i] = xh1_1[i][0]   
    canalh1_2[i] = xh1_2[i][0]   
    canalh1_3[i] = xh1_3[i][0]
    
    canali1_1[i] = xi1_1[i][0]   
    canali1_2[i] = xi1_2[i][0]   
    canali1_3[i] = xi1_3[i][0]
    
    canalo1_1[i] = xo1_1[i][0]   
    canalo1_2[i] = xo1_2[i][0]   
    canalo1_3[i] = xo1_3[i][0]
   
   
#Vector del tiempo
t = np.linspace(0,10,2000000)

#Dataframe con los canales
d = {'Tiempo': t, 'Canal 1 H-A-1': canalh1_1, 'Canal 1 H-A-2': canalh1_2,'Canal 1 H-A-3': canalh1_3,'Canal 1 I-A-1': canali1_1, 
     'Canal 1 I-A-2': canali1_2, 'Canal 1 I-A-3': canali1_3,'Canal 1 O-A-1': canalh1_1, 'Canal 1 O-A-2': canalo1_2,'Canal 1 O-A-3': canalo1_3}
canales=pd.DataFrame(d)
canales.head()
```

```python
#Guardamos cada canal en un conjunto:
scanalh1_1 = set(canalh1_1)
scanalh1_2 = set(canalh1_2)
scanalh1_3 = set(canalh1_3)

scanali1_1 = set(canali1_1)
scanali1_2 = set(canali1_2)
scanali1_3 = set(canali1_3)

scanalo1_1 = set(canalo1_1)
scanalo1_2 = set(canalo1_2)
scanalo1_3 = set(canalo1_3)

```

```python
#Indice de Jaccard canales HA1 y HA2. SIN EJECUTAR SOLO PARA EL INFORME

#1: Interseccion
inth12 = scanalh1_1 & scanalh1_2
#2: Union
unh12 = scanalh1_1 | scanalh1_2

#INDICE:
ijach12 = len(inth12)/len(unh12)
```

```python
#Comparaciones de las H
#interseccion
inth12 = scanalh1_1 & scanalh1_2
inth13 = scanalh1_1 & scanalh1_3
inth23 = scanalh1_2 & scanalh1_3

#union
unh12 = scanalh1_1 | scanalh1_2
unh13 = scanalh1_1 | scanalh1_3
unh23 = scanalh1_2 | scanalh1_3

#Indice jaccard
ijach12 = len(inth12)/len(unh12)
ijach13 =len(inth13)/len(unh13)
ijach23 =len(inth23)/len(unh23)



ijach = np.mean([ijach12, ijach13, ijach23])
```

```python
#Comparaciones de las I
#interseccion
inti12 = scanali1_1 & scanali1_2
inti13 = scanali1_1 & scanali1_3
inti23 = scanali1_2 & scanali1_3

#union
uni12 = scanali1_1 | scanali1_2
uni13 = scanali1_1 | scanali1_3
uni23 = scanali1_2 | scanali1_3

#Indice jaccard
ijaci12 = len(inti12)/len(uni12)
ijaci13 = len(inti13)/len(uni13)
ijaci23 = len(inti23)/len(uni23)




ijaci = np.mean([ijaci12, ijaci13, ijaci23])
```

```python
#Comparaciones de las O
#interseccion
into12 = scanalo1_1 & scanalo1_2
into13 = scanalo1_1 & scanalo1_3
into23 = scanalo1_2 & scanalo1_3

#union
uno12 = scanalo1_1 | scanalo1_2
uno13 = scanalo1_1 | scanalo1_3
uno23 = scanalo1_2 | scanalo1_3

#Indice jaccard
ijaco12 = len(into12)/len(uno12)
ijaco13 = len(into13)/len(uno13)
ijaco23 =len(into23)/len(uno23)



ijaco = np.mean([ijaco12, ijaco13, ijaco23])
```

```python
#Tabla de las comparaciones de los canales que son iguales

indjacc = {'Indice de jaccard': ['H', 'I', 'O'], 
          'Canal 1-2':[ijach12,ijaci12,ijaco12],
         'Canal 1-3': [ijach13,ijaci13,ijaco13],
         'Canal 2-3': [ijach23,ijaci23,ijaco23]        
         }


indicesjac = pd.DataFrame(indjacc)
indicesjac
```
